```python
import numpy as np
import pandas as pd
import matplotlib as mb
import seaborn as sns
import scipy.stats 
import requests
%matplotlib inline
```

```python

def find_df(path,ftype,label):
    data=pd.read_csv(path,header=None) 
    df_bearing=np.array(data.iloc[:,0])
    
    temp=df_bearing
    temp

    i=0
    initial_data=[]
    for i in range(0,len(temp),1670):
        initial_data.append(temp[i:i+1670])


    split=[]
    for i in initial_data:
        split.append(i[35:-35])


    mean=[]
    maximum=[]
    minimum=[]
    Variance=[]
    Range=[]
    Standard_deviation=[]
    Absolute_mean=[]
    ABSMAX=[]
    sqr=[]
    RMS=[] 
    Kurtosis=[]
    Skewness=[]
    clearance_factor=[]
    ABSV=[]

    for i in range(len(split)):
        temp_mean= np.mean(split[i])
        mean.append(temp_mean)
    #MAXIMUM
        temp_maximum= np.max(split[i])
        maximum.append(temp_maximum)
    #MINIMUM    
        temp_minimum= np.min(split[i])
        minimum.append(temp_minimum)
    #RANGE
        temp_range= np.max(split[i])- np.min(split[i])
        Range.append(temp_range)
    #STANDARD DEVIATION
        temp_std= np.std(split[i])
        Standard_deviation.append(temp_std)
    #VARIANCE
        temp_vrc= np.var(split[i])
        Variance.append(temp_vrc)
    #ABSOLUTE MEAN
        temp_abs= (np.sum(np.absolute(split[i]))/len(split[i]))
        Absolute_mean.append(temp_abs)
    #SQUARES
        temp_sqr= (np.square(split[i]))
        sqr.append(temp_sqr)
    #RMS
        temp_rms= np.sqrt(np.mean(sqr[i]))
        RMS.append(temp_rms)
    #ABSOLUTE MAXIMUM    
        temp_absmax= np.max(np.absolute(split[i]))
        ABSMAX.append(temp_absmax)
    #KURTOSIS
        temp_kur=scipy.stats.kurtosis(split[i], fisher=True)
        Kurtosis.append(temp_kur)
    #SKEWNESS
        temp_skw=scipy.stats.skew(split[i])
        Skewness.append(temp_skw)
    #ABSOLUTE VALUES  
        temp_absv=np.absolute(split[i])
        ABSV.append(temp_absv)




    for i in range(len(split)):
        #CIF
        temp_cif=ABSMAX[i]/(np.square(np.mean(np.sqrt(ABSV[i]))))
        clearance_factor.append(temp_cif)
    #clearance_factor    

    FINAL_DATA=pd.DataFrame(data=mean)
    FINAL_DATA

    headers=['Mean']
    FINAL_DATA.columns=headers
    FINAL_DATA['Maximum_value']= maximum
    FINAL_DATA['Minimum_value']= minimum
    FINAL_DATA['Range']= Range
    FINAL_DATA['Standard Deviation']=Standard_deviation
    FINAL_DATA['Variance']=Variance
    FINAL_DATA['Absolute_mean']=Absolute_mean
    FINAL_DATA['Absolute_MAX']=ABSMAX
    FINAL_DATA['RMS']=RMS
    FINAL_DATA['Crest Factor']=FINAL_DATA['Absolute_MAX']/FINAL_DATA['RMS']
    FINAL_DATA['Shape factor']=FINAL_DATA['RMS']/FINAL_DATA['Absolute_mean']
    FINAL_DATA['Impulse Factor']=FINAL_DATA['Absolute_MAX']/FINAL_DATA['Absolute_mean']
    FINAL_DATA['Kurtosis']=Kurtosis
    FINAL_DATA['Skewness']=Skewness
    FINAL_DATA['Clearance Factor']=clearance_factor
    FINAL_DATA['FAULT TYPE']=ftype
    FINAL_DATA['Label']=label        
    return FINAL_DATA
```

```python
paths_url='/Users/srinivasaadvaith/Desktop/FINAL CODES/DATA/urls/LOAD 1.csv'
paths=pd.read_csv(paths_url,header=None)
headers=['links','Fault type','Label']
paths.columns=headers
p=find_df(paths['links'][0],paths['Fault type'][0],paths['Label'][0])
for i in range(1,10): 
    s=find_df(paths['links'][i],paths['Fault type'][i],paths['Label'][i]) 
    p=pd.concat([p,s],ignore_index=True)

p.to_csv('DATA_1.csv')


#paths
```
